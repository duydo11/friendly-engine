```python
from typing import Counter
#W7A16
nums=list(map(int,input().split()))
counts=Counter(nums)
have_to_remove=0
for x,a in counts.items():
    if a<x:
      have_to_remove+=a
    elif a>x:
      have_to_remove+=(a-x)
if have_to_remove==0:
  print('YES')
else:
  print('NO',have_to_remove)
```

```python
#W7A18:
nums=list(map(int,input().split()))
chan=[]
le=[]
for num in nums:
  if num%2==0:
    chan.append(num)
  else:
    le.append(num)
chan.sort()
le.sort(reverse=True)
result=chan+le
print(*result)
```

```python
#W7A19:
line1=list(map(int,input().split()))
line2=list(map(int,input().split()))
result=line1+line2
for i in range(len(line1)-1):
  if line1[i]<=line1[i+1]:
    result.sort()
  elif line1[i]>=line1[i+1]:
    result.sort(reverse=True)
print(*(result))
```

```python
#W7A20:
def demsocs(n):
  count=0
  while n>0:
    n=n//10
    count+=1
  return count
nums=list(map(int,input().split()))
n=int(input())
tong=0
for i in range(len(nums)):
  if demsocs(nums[i])==n:
    tong+=nums[i]
print(tong)
```

```python
#W7A22:
from collections import Counter
n=int(input())
chuoi=[]
ten=[]
names=[]
for i in range(n*2):
  name=input()
  chuoi.append(name)
for i in range(0,len(chuoi),2):
  last_name=chuoi[i]
  first_name=chuoi[i+1]
  full_name=last_name+' '+first_name
  ten.append(full_name)
  names.append(first_name)
names.sort()
result=names
count=Counter(result)
print(ten,end='\n')
print(max(count,key=count.get))
```

```python
#W7A25:
s=input().strip()
n=list(s)
k=int(input())
ket_qua=[]
for _ in range(len(n)-k):
  i=0
  while i<(len(n)-1) and (n[i]>=n[i+1]):
    i+=1
  n.pop(i)
ket_qua.append(n)
for i in range(len(ket_qua)):
  ket_qua[i]=''.join(ket_qua[i])
print(*ket_qua)
```

```python
#W7A28:
def ktra_snt(n):
  if n<2:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True

snt=[]
n=int(input())
for i in range(n+1):
  if ktra_snt(i):
    snt.append(i)
if n not in snt:
  snt.append(n)

max_gap=0
for i in range(len(snt)-1):
  gap=snt[i+1]-snt[i]-1
  if gap>max_gap:
    max_gap=gap

for i in range(len(snt)-1):
  gap=snt[i+1]-snt[i]-1
  if gap==max_gap:
    result=list(range(snt[i]+1,snt[i+1]))
    print(result)
    break
```

```python
#W7A29
def doan_tang_dan(danhsach):
  max_len=1
  current_len=1
  start=0
  current=0
  for i in range(1,len(danhsach)):
    if danhsach[i]>danhsach[i-1]:
      current_len+=1
    else:
      if current_len>max_len:
        max_len=current_len
        start=current
      current_len=1
      current=i
    if current_len>max_len:
      max_len=current_len
      start=current
  return danhsach[start:start+max_len]
danhsach=list(map(int,input().split()))
print(doan_tang_dan(danhsach))
```
