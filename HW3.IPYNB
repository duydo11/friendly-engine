```python
#6
def uoc(n):
    for i in range (1 , int ( n ** 0.5) + 1) :
        if n % i == 0  and i * i != n:
            print (i , end = ' ')
            print (n // i , end = ' ')
        elif i * i == n:
            print (i , end = ' ')
n = int ( input ())
uoc(n)
```

```python
#7
def demuoc (n):
    dem = 0
    for i in range (1 , int ( n ** 0.5) + 1) :
        if n % i == 0  and i * i != n:
            dem += 2
        elif i * i == n:
            dem += 1
    return dem
n = int ( input ())
print (demuoc(n))
```

```python
#8
def uocchung (a , b):
    for i in range (1 , min (a , b) + 1):
        if a % i == 0 and b % i == 0:
            print (i , end = ' ')
a , b = map ( int , input().split())
uocchung (a , b)
```

```python
#9
def scp (n) :
    k = int ( n ** 0.5)
    if k * k == n :
        return True
    else :
        return False
n = int ( input ())
if scp (n) :
    print (n ,"la so chinh phuong")
else :
    print (n ,"khong la so chinh phuong")
```

```python
#10
def tonguoc (n) :
    sum = 0 
    for i in range (1 , int (n ** 0.5)+1) :
        if n % i == 0 and i * i != n :
            sum += i + n // i
        elif i * i == n :
            sum += i
    return sum-n
n = int ( input ())
if tonguoc (n) == n :
    print (n , "la so hoan hao")
else :
    print (n , "khong la so hoan hao")
```
