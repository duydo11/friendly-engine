```python
#W11A4:
import re

def clean_string(text):
    text = text.lower()
    text = re.sub(r'[^a-z]', '', text)
    return text

def areAnagrams(path1: str, path2: str) -> bool:
    try:
        with open(path1, 'r', encoding='utf-8') as f1, \
             open(path2, 'r', encoding='utf-8') as f2:
            str1 = f1.read()
            str2 = f2.read()

        clean_str1 = clean_string(str1)
        clean_str2 = clean_string(str2)


        if len(clean_str1) != len(clean_str2):
            return False
        return sorted(clean_str1) == sorted(clean_str2)

    except FileNotFoundError:
        print("Lỗi: Không tìm thấy file.")
        return False
print(areAnagrams('file1.txt', 'file2.txt'))
```

```python
#W11A8:
def write_fibonacci(path):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            n = int(f.readline().split()[0])
            a = int(f.readline().split()[0])
            b = int(f.readline().split()[0])
    except FileNotFoundError:
        print("Không tìm thấy tệp tin!")
        return

    fib = [0] * (n + 1)
    if n >= 0: fib[0] = a
    if n >= 1: fib[1] = b

    for i in range(2, n + 1):
        fib[i] = fib[i-1] + fib[i-2]

    try:
        with open('output.txt', 'w', encoding='utf-8') as f:
            for num in fib:
                f.write(str(num) + "\n")
        print("Đã ghi file thành công.")
    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

write_fibonacci('input.txt')
with open('output.txt', 'r', encoding='utf-8') as f:
  content = f.read()
  print(content)
```

```python
#W11A10:
def getMoney(path: str) -> list[int]:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            content = f.read()
            money_list = [int(x) for x in content.split()]
            return money_list
    except FileNotFoundError:
        print("Không tìm thấy tệp tin!")
        return []
    except Exception as e:
        print(f"Lỗi: {e}")
        return []

def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]

    prev2 = 0
    prev1 = 0

    for amount in houses:
        current = max(prev1, prev2 + amount)

        prev2 = prev1
        prev1 = current

    return prev1

with open('money.txt', 'w') as f:
    f.write("10 20 30 1 5")

money_in_houses = getMoney('money.txt')
max_money = rob(money_in_houses)

print(f"Danh sách số tiền trong các nhà: {money_in_houses}")
print(f"Số tiền tối đa có thể trộm được: {max_money}")
```

```python
#W11A12:
def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r') as f:
            data = f.read().split()
            numbers = [int(x) for x in data[1:]]

            numbers.sort()

            option1 = numbers[-1] * numbers[-2] * numbers[-3]
            option2 = numbers[0] * numbers[1] * numbers[-1]

            return max(option1, option2)
    except:
        return 0
print(maximumProduct('input.txt'))
```

```python
#W11A13:
def moveZeroes(path: str):
    try:
        with open(path, 'r') as f:
            n_str = f.readline().strip()
            n = int(n_str)

            line_str = f.readline().strip()
            nums = [int(x) for x in line_str.split()]

        non_zeroes = [x for x in nums if x != 0]
        zeroes_count = nums.count(0)

        result = non_zeroes + [0] * zeroes_count
        return result

    except FileNotFoundError:
        print("Không tìm thấy tệp tin!")
        return 0
    except ValueError:
        return 0
print(moveZeroes('/content/input.txt'))
```

```python
#W11A14:
def averageTime(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        if len(lines) <= 1:
            return 0.0

        total_time = 0.0
        count = 0

        for line in lines[1:]:
            line = line.strip()
            if not line:
                continue

            parts = line.split(',')

            time_value = float(parts[1])
            total_time += time_value
            count += 1

        if count == 0:
            return 0.0

        return round(total_time / count, 2)

    except (FileNotFoundError, ValueError, IndexError):
        return 0
print(averageTime('du_lieu_dua.csv'))
```

```python
#W11A16:
def shape(path:str):
  try:
    with open(path, 'r', encoding = 'utf-8') as f:
      lines = f.readlines()
    results = []
    for line in lines[1:]:
      line = line.strip()
      if not line:
        continue

      parts = line.split(',')
      if len(parts) >= 4:
        area = 0.0
        shape_type = parts[0].upper()

        if shape_type == 'SQUARE':
          area = float(parts[1])**2
        elif shape_type == 'RECTANGLE':
          dims = parts[2].split()
          if len(dims) >= 2:
            area = float(dims[0]) * float(dims[1])
        elif shape_type == 'CIRCLE':
          area = 3.14 * float(parts[3])**2
        results.append(f'{area:.2f}')

    return results
  except FileNotFoundError:
    print("Không tìm thấy tệp tin!")
shape('/content/shape.csv')
```

```python
#W11A17:
def productExceptSelf(path: str) -> list[int]:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        if len(lines) < 2:
            return []

        data_string = " ".join(lines[1:]).strip()
        nums = [int(x) for x in data_string.replace(',', ' ').split()]

        n = len(nums)
        if n <= 1:
            return []

        res = [1] * n

        prefix = 1
        for i in range(n):
            res[i] = prefix
            prefix *= nums[i]

        suffix = 1
        for i in range(n - 1, -1, -1):
            res[i] *= suffix
            suffix *= nums[i]

        print(res)
        return res

    except Exception as e:
        print(f"Lỗi: {e}")
        return []

result = productExceptSelf('/content/data.txt')
```

```python
#W11A18
import math

def estimatedTime(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        src_parts = lines[0].split()
        x1, y1 = float(src_parts[2]), float(src_parts[4])

        des_parts = lines[1].split()
        x2, y2 = float(des_parts[2]), float(des_parts[4])

        vel_parts = lines[2].split()
        v = float(vel_parts[1])

        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        time = distance / v

        return round(time, 2)

    except Exception as e:
        print(f"Lỗi xử lý tệp: {e}")
        return 0.0

print(estimatedTime('/content/missile_config.txt'))
```

```python
#W11A19
import struct

def read_binary_doubles():
    path = input("Nhập đường dẫn đến tệp nhị phân: ")
    try:
        with open(path, 'rb') as f:
            content = f.read()
            count = len(content) // 8
            numbers = struct.unpack('d' * count, content)

            for num in numbers:
                print(f"{num:.5f}")
    except FileNotFoundError:
        print("Không tìm thấy tệp!")
```
