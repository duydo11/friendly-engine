```python
#W8A1:
class Cylinder:
  def __init__(self,r,h):
    self.r=r
    self.h=h
  def surface_area(self):
    return 2*3.14*(self.r**2) + self.r*self.h
  def volume(self):
    return 3.14*(self.r**2)*self.h
r,h=map(float,input().split())
c=Cylinder(r,h)
print(f'{c.surface_area():.2f}')
print(f'{c.volume():.2f}')
```

```python
#W8A2:
class Date:
  def __init__(self,d,m,y):
    self.d=d
    self.m=m
    self.y=y

  def kiemtra(self):
    days_in_month=[0,31,28,31,30,31,30,31,31,30,31,30,31]
    if self.y%4==0 and self.y%100!=0 or self.y%400==0:
      days_in_month[2]=29
    if self.m<1 or self.m>12 or self.d>days_in_month[self.m] or self.d<1:
      return 'INVALID'
    else:
      if self.d==days_in_month[self.m]:
        self.d=1
        self.m+=1
        if self.d==1 and self.m==13:
          self.d=1
          self.m=1
          self.y+=1
      return f'{self.d}/{self.m}/{self.y}'

d,m,y=map(int,input().split('/'))
date=Date(d,m,y)
print(date.kiemtra())
```

```python
#W8A3:
import math
class Point:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def check(self):
    if self.x==0 and self.y==0:
      return 'Gốc tọa độ'
    elif self.x==0 and self.y!=0:
      return 'Trục tung'
    elif self.y==0 and self.x!=0:
      return 'Trục hoành'
    else:
      return 'Không thuộc trục tung, trục hoành'
  def distance(self,other):
    return math.sqrt((self.x-other.x)**2 + (self.y-other.y)**2)
x1,y1=map(float,input().split())
M=Point(x1,y1)
x2,y2=map(float,input().split())
O=Point(x2,y2)
print(M.check())
print(f'{M.distance(O):.2f}')
```

```python
#W8A6:
class Rectangle:
  def __init__(self,w,h,k):
    self.w=w
    self.h=h
    self.k=k
  def area(self):
    return self.w * self.h
  def perimeter(self):
    return 2 * (self.w + self.h)
  def scale(self,k):
    self.w = k * self.w
    self.h = k * self.h
w,h,k=map(float,input().split())
hcn=Rectangle(w,h,k)
hcn.scale(k)
print(f'{hcn.area():.2f} {hcn.perimeter():.2f}')
```

```python
#W8A7:
import math
class Fraction:
  def __init__(self,num,den):
    if den<0:
      num = -num
      den = -den
    common = math.gcd(num,den)
    self.num = num//common
    self.den = den//common
  def __str__(self):
    return f'{self.num}/{self.den}'
  def add(self,other):
    new_num = self.num * other.den + self.den * other.num
    new_den = self.den * other.den
    return Fraction(new_num,new_den)
  def subtract(self,other):
    new_num = self.num * other.den - self.den * other.num
    new_den = self.den * other.den
    return Fraction(new_num,new_den)
  def multiply(self,other):
    new_num = self.num * other.num
    new_den = self.den * other.den
    return Fraction(new_num,new_den)
  def divide(self,other):
    new_num = self.num * other.den
    new_den = self.den * other.num
    return Fraction(new_num,new_den)
line=input().split()
if len(line)==5:
  a=int(line[0])
  b=int(line[1])
  op=line[2]
  c=int(line[3])
  d=int(line[4])
n1=Fraction(a,b)
n2=Fraction(c,d)
if op=='+':
  print(n1.add(n2))
elif op=='-':
  print(n1.subtract(n2))
elif op=='*':
  print(n1.multiply(n2))
elif op=='/':
  print(n1.divide(n2))
```

```python
#W8A8:
class BankAccount:
  def __init__(self,owner,balance):
    self.owner = owner
    self.balance = balance
  def DEPOSIT(self,x):
      self.balance+=x
  def WITHDRAW(self,x):
    if x<=self.balance:
      self.balance-=x
    elif x>self.balance:
      return 'Bỏ qua giao dịch'
  def __str__(self):
    return f'{self.owner}:{self.balance:.2f}'

line=input().split()
name=line[0]
initial_balance=float(line[1])
ba = BankAccount(name,initial_balance)
n = int(input())
for _ in range(n):
  command_line=input().split()
  cmd=command_line[0]
  x=float(command_line[1])
  if cmd == 'DEPOSIT':
    ba.DEPOSIT(x)
  elif cmd == 'WITHDRAW':
    ba.WITHDRAW(x)
print(ba)

```

```python
#W8A9:
class Student:
  def __init__(self,name):
    self.name=name
    self.scores=[]
  def add_score(self,subject,score):
    self.scores.append(score)
  def avg(self):
    if not self.scores:
      return 0
    return sum(self.scores)/len(self.scores)
  def rank(self):
    a = self.avg()
    if a>=8:
      return 'Good'
    elif 6.5<=a<8:
      return 'Average'
    elif 5<=a<6.5:
      return 'Poor'
name=input()
n=int(input())
student=Student(name)
for _ in range(n):
  sub_sc=input().split()
  subject=sub_sc[0]
  score=float(sub_sc[1])
  student.add_score(subject,score)
print(f'{student.avg():.2f}')
print(student.rank())
```

```python
#W8A10:
class Book:
  def __init__(self,title,author,year):
    self.title=title
    self.author=author
    self.year=year

class Library:
  def __init__(self):
    self.books=[]
  def add(self,book):
    self.books.append(book)
  def count_by_author(self,name):
    count = sum(1 for book in self.books if book.author==name)
    return count
  def find_by_year(self,y):
    count = sum(1 for book in self.books if book.year==y)
    return count

n=int(input('Nhập số thao tác'))
lib=Library()

for _ in range(n):
  line=input().split()
  cmd=line[0]
  info=line[1]

  if cmd =='ADD':
    title=info.split(';')[0]
    author=info.split(';')[1]
    year=int(info.split(';')[2])
    book=Book(title,author,year)
    lib.add(book)
  elif cmd == 'COUNT':
    print(lib.count_by_author(info))
  elif cmd == 'COUNTYEAR':
    print(lib.find_by_year(int(info)))



```

```python
#W8A11:
import math
class Point:
  def __init__(self,x,y,z):
    self.x = x
    self.y = y
    self.z = z
  def __str__(self):
    return f'({self.x},{self.y},{self.z})'
  def vector(self,other):
    new_x = self.x - other.x
    new_y = self.y - other.y
    new_z = self.z - other.z
    return Point(new_x,new_y,new_z)
  def vector_phaptuyen(self,other):
    new_x = self.y * other.z - other.y * self.z
    new_y = self.z * other.x - other.z * self.x
    new_z = self.x * other.y - other.x * self.y
    return Point(new_x,new_y,new_z)
  def cos_phi(self,other):
    a = abs(self.x * other.x + self.y * other.y + self.z * other.z)
    b = math.sqrt(self.x**2 + self.y**2 + self.z**2)
    c = math.sqrt(other.x**2 + other.y**2 + other.z**2)
    return a/(b*c)

xA,yA,zA = map(float,input().split())
xB,yB,zB = map(float,input().split())
xC,yC,zC = map(float,input().split())
xD,yD,zD = map(float,input().split())

A = Point(xA,yA,zA)
B = Point(xB,yB,zB)
C = Point(xC,yC,zC)
D = Point(xD,yD,zD)

u1=A.vector(B)
u2=B.vector(C)
u3=C.vector(D)

n1=u1.vector_phaptuyen(u2)
n2=u2.vector_phaptuyen(u3)

print(f'{math.degrees(math.acos(n1.cos_phi(n2))):.2f}')
```

```python
#W8A12:
class ComplexNumber:
  def __init__(self,r,i):
    self.r=r
    self.i=i
  def multiply(self,other):
    new_r = (self.r * other.r) -(self.i * other.i)
    new_i = (self.r * other.i) + (self.i * other.r)
    return ComplexNumber(new_r,new_i)
  def __str__(self):
    return f'{self.r} {self.i}'
r1,i1 = map(float,input().split())
complex1 = ComplexNumber(r1,i1)
r2,i2 = map(float,input().split())
complex2 = ComplexNumber(r2,i2)
result = complex1.multiply(complex2)
print(f'{result.r} {result.i}')
```
