```python
#w7a1
def binary_search(n,x) :
    r  = len(n)-1 
    l = 0
    while l < r :
        m = int ((r+ l)/2)
        if n[m] == x :
            return m 
        elif n[m]  < x :
            l = m +1
        elif n[m] > x :
            r = m -1
    return -1
while True :
    a = list( map (int , input().split()))
    k = int (input())
    print(binary_search(a,k))

        
```

```python
#w7a3
def bubble_sort( a) :
    for i in range ( len(a)) :
        for j in range( i , len(a)-1) :
            if a[j] > a[j+1] :
                a[j] , a[j+1] = a[j+1] , a[j] 
    return a 
def selection_sort(a) :
    for i in range (len(a)) :
        min = i 
        for j in range ( i , len(a)) :
            if a[min] > a[j] :
                min = j
        a[i] , a[min] = a[min] , a[i]
    return a
a = list(map (int , input ().split()))
print( bubble_sort(a))
print( selection_sort(a))
```

```python
#w7a7
#QHD
#D[] : dãy lưu trữ độ dài của các phânf tử a
#D[i] : độ dài của dãy con dài nhất kết thúc ở chỉ số i

def maxx( a , b) :
    x = int (a) 
    y = int (b)
    if x > y :
        return x
    elif y > x :
        return y
    elif x== y :
        return y
def check(a) :
   
    ans = 0
    D=[]
    for i in range (0,len(a)) :
        D.append(1)
    for i in range (0, len(a)) :
        # duyet qua các phan tu truoc i
        for j in range ( 0, i) :
            if a[i] > a[j] :
                k = D[j]+1
                D[i] = maxx(D[i],k)
    for i in D :
        if i > ans :
            ans = i 
    return ans
a = list ( map (int , input().split()))
print(check(a))
```
