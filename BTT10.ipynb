```python
def so_dac_biet(n):
  luc_sau = []
  while n > 0:
    digit = n%10
    luc_sau.append(digit)
    n = n//10
  check = all(luc_sau[i-1] == luc_sau[i] for i in range(1,len(luc_sau)))
  return check

t = int(input())
nums = []
for i in range(t):
  num = int(input())
  sdb = []
  for j in range(1,num+1):
    if so_dac_biet(j):
      sdb.append(j)
  nums.append(len(sdb))
print(nums)
```
